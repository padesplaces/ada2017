```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url_genres = "http://www.musicgenreslist.com"
```

```python
genres_r = r = requests.get(url_genres)
```

```python
soup = BeautifulSoup(genres_r.text, "html.parser")
```

```python
caca = soup.find('div', class_="one_half")
soup2 = BeautifulSoup(str(caca), "html.parser")
```

```python
all_categories = soup2.find_all("li")
```

```python
def remove_useless_spaces(x):
    clean = x
    if clean[0] == " ":
        clean = clean[1:]
    if clean[-1] == " ":
        clean = clean[0:-1]
    return clean
        
```

```python
dict_cat = dict()
all_genres = list()
general_genres = list()
for cat in all_categories:
    soup_cat = BeautifulSoup(str(cat), "html.parser")
    if soup_cat.find("ul") != None:
        a_ref = soup_cat.find("a")
        if a_ref == None:
            title = soup_cat.find("li").text
        else : 
            title = a_ref.text
        if "(" in title:
            title = title.split("(")[0]
        if "/" in title:
            split_title1 = title.split("/")[0]
            split_title2 = title.split("/")[1]
            general_genres.append(remove_useless_spaces(split_title1))
            general_genres.append(remove_useless_spaces(split_title2))
        elif " – " in title:
            if "&" in title.split(" – ")[1]:
                split_title1 = title.split(" – ")[1].split("&")[0]
                split_title2 = title.split(" – ")[1].split("&")[1]
                general_genres.append(remove_useless_spaces(split_title1))
                general_genres.append(remove_useless_spaces(split_title2))
        else:
            general_genres.append(remove_useless_spaces(title))
        subcat = soup_cat.find_all("ul")
        soup_subcat = BeautifulSoup(str(subcat), "html.parser")
        list_subcat = soup_subcat.find_all("li")
        list_subcats = list()
        for s in list_subcat:
            category = s.text
            if "\xa0" in category:
                category = category.replace("\xa0", " ")
            if "(" in category:
                category = category.split("(")[0]
            if "/" in category :
                category1 = remove_useless_spaces(category.split("/")[0])
                category2 = remove_useless_spaces(category.split("/")[1])
                
                list_subcats.append(category1)
                list_subcats.append(category2)
                all_genres.append(category1)
                all_genres.append(category2)
            else:
                all_genres.append(remove_useless_spaces(category))
                list_subcats.append(remove_useless_spaces(category))
        dict_cat[title] = list_subcats
```
