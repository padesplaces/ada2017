```python
import pickle
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pickle.load(open("covers_final.p","rb"))
df
```

```python
df.index.levels[0].size
```

```python
zero_pad = pd.Series(0, index=np.arange(1910,2016))
zero_pad.rename("year",inplace=True)
zero_pad
```

## FIRST PLOT : originals

```python
originals = df[df.status == "original"]
originals
```

```python
years_originals = originals.year.dropna().astype(int).value_counts().sort_index()
years_originals
```

```python
years_originals_padded = zero_pad.add(years_originals,fill_value=0).astype(int)
years_originals_padded
```

```python
years_originals_padded.plot(kind="bar")
plt.xticks(np.arange(0,len(years_originals_padded),10), np.arange(years_originals_padded.keys()[0],years_originals_padded.keys()[-1],10),rotation=0)
plt.show()
```

## SECOND PLOT : covers

```python
covers = df[df.status == "cover"]
covers
```

```python
years_covers = covers.year.dropna().astype(int).value_counts().sort_index()
years_covers
```

```python
years_covers_padded = zero_pad.add(years_covers,fill_value=0).astype(int)
years_covers_padded
```

```python
years_covers_padded.plot(kind="bar")
plt.xticks(np.arange(0,len(years_covers_padded),10), np.arange(years_covers_padded.keys()[0],years_covers_padded.keys()[-1],10),rotation=0)
plt.show()
```

## THIRD PLOT

```python
df_valid_year = df[df.year.notnull()]
```

```python
df_valid_year.year.astype(int).min()
```

```python
clique_size = df.groupby("clique_id").size()
```

```python
covers_per_original_year = []

for year in range(1910,2015):
    originals_by_year = df_valid_year[(df_valid_year.status == "original") & (df_valid_year.year.astype(int) == year)]
    id_cliques_original = originals_by_year.index.get_level_values(0).tolist()
    nb_covers = clique_size[id_cliques_original].sum()
    covers_per_original_year.append(nb_covers)

covers_per_original_year
```

```python
len(covers_per_original_year)
```

```python
plt.bar(np.arange(len(covers_per_original_year)),covers_per_original_year)
plt.xticks(np.arange(0,len(covers_per_original_year),10), np.arange(1910,2015,10),rotation=0)
plt.show()
```
