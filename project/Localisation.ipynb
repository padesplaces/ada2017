```python
def get_artist_location(artist):
    r = requests.get('https://secondhandsongs.com/search?search_text='+artist)
    soup = BeautifulSoup(r.text, 'html.parser')
    title = soup.find('h2', class_="section-title")
    table = soup.find('table', class_='table data object table-striped')
    # Check if the result is an artist
    artist_id = ""
    try:
        if "Artists" in title.text:
        # Get the artist ID to find the its web page on SHS
            # When only one artist is found
            if "1+" in title.text:
                artist_id = str(table.a).split("\"")[1]
            # When several artists are found
            else:
                artists = table.find_all("a")
                for art in artists:
                    if art.text != "\n\n":
                        if artist == art.text:
                            artist_id = str(art).split("\"")[1]
                            break

            r_artist = requests.get('https://secondhandsongs.com'+artist_id)
            soup2 = BeautifulSoup(r_artist.text, 'html.parser')
            country = soup2.find("dd", itemprop="nationality")
            return country.text
        else:
            return "Unavailable"
    except:
        print(artist)
```

```python
shs_df["Country"] = shs_df["name"].map(lambda x : get_artist_location(x))
```
